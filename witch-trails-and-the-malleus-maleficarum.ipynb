Note: Technically, Germany did not exist until the 1800s, and for a majority of the timeline was the Holy Roman Empire; however, "German" and "Germany" will be used in this notebook for brevity.

```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
import os
for dirname, _, filenames in os.walk('/kaggle/input'):
    for filename in filenames:
        print(os.path.join(dirname, filename))
        
print('Complete!')
```

```python
witches = pd.read_csv('../input/witch-trials/trials.csv')
witches.head()
```

```python
witchtable = pd.pivot_table(witches, values=['tried','deaths'], index=['decade'],aggfunc=np.sum)
plt.plot(witchtable)
plt.legend()
plt.title("Witch Trials and Executions, by Decade, from 1300 to 1850")
plt.show()
witchtable2 = pd.pivot_table(witches, values=['tried','deaths'], index=['century'],aggfunc=np.sum)
plt.plot(witchtable2)
plt.legend()
plt.title("Witch Trials and Executions, by Century, from 1300 to 1850")
plt.show()
```

```python
germany = witches[witches['gadm.adm0'] == 'Germany']
germantable = pd.pivot_table(germany, values=['tried','deaths'], index=['decade'],aggfunc=np.sum)
plt.plot(germantable)
plt.legend()
plt.title("Witch Trials and Executions in Germany, by Decade, from 1300 to 1790")
plt.show()
germantable2 = pd.pivot_table(germany, values=['tried','deaths'], index=['century'],aggfunc=np.sum)
plt.plot(germantable2)
plt.legend()
plt.title("Witch Trials and Executions in Germany, by Century, from 1300 to 1700")
plt.show()
```

Looking at trials and deaths as a whole, as well as in Germany alone (where the Malleus Maleficarum originated), it doesn't look like there was an immediate effect on hunting witches. In fact, looking at Germany alone, the decade in which the most people were tried and killed for being witches were the 1620s, towards the beginning of the Thirty Years' War, which was fought heavily along religious lines (and led to the deaths of roughly 1 in 3 Germans). 

If anyone wants the quick wikipedia link to the Thirty Years' War, here it is: https://en.wikipedia.org/wiki/Thirty_Years%27_War.

Interestingly, following the war (which ended in 1648), there is a sharp decline in the number of deaths, especially in the ratio of deaths to trials. Perhaps all of the bloodshed, death, and distruction made the Germans less interested in killing, as well as the movements towards religious tolerance from the Treaty of Westphalia.
